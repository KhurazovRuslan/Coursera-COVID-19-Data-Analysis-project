```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# inline plotting and style
%matplotlib inline
sns.set_style('darkgrid')
```

```python
# importing dataset
corona = pd.read_csv('covid19_Confirmed_dataset.csv')
# head of the dataframe
corona.head()
```

```python
# dataset's shape
corona.shape
```

```python
# dataset's info
corona.info()
```

```python
# we don't need Lat and Long columns
corona.drop(['Lat','Long'], axis=1, inplace=True)
```

```python
# we need data to be grouped by Country/Region
corona_aggregated = corona.groupby('Country/Region').sum()
corona_aggregated.head(10)
```

```python
# shape and info of our new dataset
corona_aggregated.shape
```

```python
corona_aggregated.info()
```

```python
# let's visualize a cummulative number of confirmed cases in different countries. E.g. China, Russia, Italy and Spain
corona_aggregated.loc['China'].plot()
corona_aggregated.loc['Russia'].plot()
corona_aggregated.loc['Italy'].plot()
corona_aggregated.loc['Spain'].plot()
plt.legend()
```

```python
# let's plot number of cases for China first 3 days
corona_aggregated.loc['China'][:3].plot()
```

```python
# let's plot the infection rate by day for Russia
corona_aggregated.loc['Russia'].diff().plot()
```

```python
# let's find max infection rate for China, Russia, Italy and Spain
print('Max infection rate in China was',corona_aggregated.loc['China'].diff().max())
print('Max infection rate in Russia was',corona_aggregated.loc['Russia'].diff().max())
print('Max infection rate in Italy was',corona_aggregated.loc['Italy'].diff().max())
print('Max infection rate in Spain was',corona_aggregated.loc['Spain'].diff().max())
```

```python
# let's find max infection rate for all countries in the dataset
countries = list(corona_aggregated.index)
max_infection_rate = []
for country in countries:
    max_infection_rate.append(corona_aggregated.loc[country].diff().max())
```

```python
# let's put max_infection_rate in our dataset
corona_aggregated['max_infection_rate'] = max_infection_rate
```

```python
# let's create a new daytaframe with max_infectioni_rate
corona_infection_rate = pd.DataFrame(corona_aggregated['max_infection_rate'])
corona_infection_rate.head()
```

```python
# importing happiness report
happiness = pd.read_csv('worldwide_happiness_report.csv')
happiness.head()
```

```python
# let's drop the columns we're not gonna use
happiness.drop(['Overall rank','Score','Generosity','Perceptions of corruption'], axis=1, inplace=True)
```

```python
happiness.head()
```

```python
# let's change indices of the dataset to Country or region
happiness.set_index('Country or region', inplace=True)
happiness.head()
```

```python
# now let's create a new dataframe joining prepared datasets
df = corona_infection_rate.join(happiness, how='inner')
df.head()
```

```python
# correlation matrix
df.corr()
```

```python
# let's plot the correlations between our columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16,10), sharey=True)
sns.regplot(x=df['GDP per capita'], y=np.log(df['max_infection_rate']), ax=axes[0][0])
axes[0][0].set_title('GDP vs Max infection rate')
sns.regplot(x=df['Social support'], y=np.log(df['max_infection_rate']), ax=axes[0][1])
axes[0][1].set_title('Social support vs Max infection rate')
sns.regplot(x=df['Healthy life expectancy'], y=np.log(df['max_infection_rate']), ax=axes[1][0])
axes[1][0].set_title('Healthy life expectancy vs Max infection rate')
sns.regplot(x=df['Freedom to make life choices'], y=np.log(df['max_infection_rate']), ax=axes[1][1])
axes[1][1].set_title('Freedom to make life choices vs Max infection rate')
plt.tight_layout()
```
